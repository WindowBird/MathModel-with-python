```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# 已知数据点
x = np.linspace(0, 10, num=11, endpoint=True)
y = np.cos(-x**2/9.0)

x,y
```

```python
# 创建一个线性插值函数
f = interp1d(x, y)

# 创建一个新的输入范围
xnew = np.linspace(0, 10, num=41, endpoint=True)

# 画图
plt.plot(x, y, 'o', xnew, f(xnew), '-')
plt.legend(['data', 'linear'], loc='best')
plt.show()
```

```python
import numpy as np
from scipy.interpolate import griddata
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm

# 随机生成散点数据
np.random.seed(0)
x = np.random.randint(-100, 100, size=100)
y = np.random.randint(-100, 100, size=100)
z = np.random.randint(-100, 100, size=100)

# 创建网格
xi = np.linspace(-100, 100, 1000)
yi = np.linspace(-100, 100, 1000)
xi, yi = np.meshgrid(xi, yi)

# 使用cubic方法进行插值
zi = griddata((x, y), z, (xi, yi), method='cubic')

# 绘制插值后的结果
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(xi, yi, zi, cmap=cm.viridis,
                       linewidth=0, antialiased=True)
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
