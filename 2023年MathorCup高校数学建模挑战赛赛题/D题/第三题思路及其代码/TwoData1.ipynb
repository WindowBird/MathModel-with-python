```python
import pandas as pd

# 将指定的一列数据转换为独热编码
column_name = 'ARN（机号）'  # 需要转换的列名
df_one_hot = pd.get_dummies(df[column_name], prefix=column_name)  # 进行独热编码
df = pd.concat([df, df_one_hot], axis=1)  # 将转换后的独热编码数据合并到原始数据框中
df.drop(column_name, axis=1, inplace=True)  # 删除原始数据列
```
