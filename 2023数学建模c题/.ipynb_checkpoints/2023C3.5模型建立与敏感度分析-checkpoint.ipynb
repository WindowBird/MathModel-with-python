```python
# Function to run the optimization
def run_optimization(intercepts, slopes, predicted_sales):
    def objective(x):
        sales = x
        markup = intercepts + slopes * sales
        return -np.sum(sales * markup)

    def constraint1(x):
        epsilon = 1e-5
        sales = x
        markup = intercepts + slopes * sales
        return epsilon - np.min(markup)

    def constraint2(x):
        sales = x
        return 1.3 * predicted_sales - sales

    def constraint3(x):
        sales = x
        return sales - 0.7 * predicted_sales
    
    bounds_sales = [(0.7 * pred_sale, 1.3 * pred_sale) for pred_sale in predicted_sales]
    x0 = np.array(predicted_sales)
    con1 = {'type': 'ineq', 'fun': constraint1}
    con2 = {'type': 'ineq', 'fun': constraint2}
    con3 = {'type': 'ineq', 'fun': constraint3}
    
    res = minimize(objective, x0, constraints=[con1, con2, con3], bounds=bounds_sales)
    
    optimized_sales = res.x if res.success else None
    optimized_value = -res.fun if res.success else None
    
    return optimized_sales, optimized_value, res.message

# Sensitivity analysis: Vary intercepts and slopes by +/- 5%, 10%, 15%
sensitivity_results = {}
for percentage in [5, 10, 15]:
    for param in ['Intercept', 'Slope']:
        for direction in ['increase', 'decrease']:
            factor = 1 + percentage / 100 if direction == 'increase' else 1 - percentage / 100
            new_params = intercepts if param == 'Intercept' else slopes
            new_params = new_params * factor

            if param == 'Intercept':
                optimized_sales, optimized_value, message = run_optimization(new_params, slopes, predicted_sales)
            else:
                optimized_sales, optimized_value, message = run_optimization(intercepts, new_params, predicted_sales)
            
            key = f"{param} {direction} by {percentage}%"
            sensitivity_results[key] = {
                'Optimized Sales': optimized_sales,
                'Optimized Value': optimized_value,
                'Status': message
            }

sensitivity_results.keys()  # Display keys to give an overview of the results
```
