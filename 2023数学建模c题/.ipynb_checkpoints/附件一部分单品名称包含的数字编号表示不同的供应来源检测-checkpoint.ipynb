```python
import pandas as pd

# Load data
xls1 = pd.ExcelFile("E:\\数学建模国赛2023选题\\C题\\附件1.xlsx")
df1=pd.read_excel(xls1)
df1
```

```python
# Check if any of the '单品名称' contains numerical identifiers indicating different supply sources.
# We'll use a regular expression to find entries that contain numbers.

import re

# Extracting the unique single product names
unique_product_names = df1['单品名称'].unique()

# Finding product names that contain numerical identifiers
products_with_numbers = [name for name in unique_product_names if re.search(r'\d', name)]

products_with_numbers[:250], len(products_with_numbers)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Rename the DataFrame to df1 as requested


# Extract the root name of each product by removing the numerical identifiers
df1['RootProductName'] = df1['单品名称'].apply(lambda x: re.sub(r'\(\d+\)', '', x))
```

```python
df1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
plt.rcParams['font.sans-serif'] = ['SimHei']


# Extract the base names of the products (i.e., names without the numerical identifiers) for those 64 items
base_product_names = [re.sub(r'\(\d+\)', '', name) for name in products_with_numbers]

# Count the number of supply sources for each base product name
supply_source_counts = Counter(base_product_names)

# Convert the counter object to a DataFrame for visualization
df_supply_sources = pd.DataFrame.from_dict(supply_source_counts, orient='index', columns=['供货来源数']).reset_index()
df_supply_sources.rename(columns={'index': '单品名称'}, inplace=True)

# Sort the DataFrame by the number of supply sources
df_supply_sources.sort_values(by='供货来源数', ascending=False, inplace=True)

# Visualize the data
plt.figure(figsize=(5, 5.5))
sns.barplot(y='单品名称', x='供货来源数', data=df_supply_sources, orient='h')
plt.title('单品名称与供货来源数')
plt.xlabel('供货来源数')
plt.ylabel('单品名称')
plt.show()

```
