```python
import pandas as pd

# Load data
xls1 = pd.ExcelFile("E:\\数学建模国赛2023选题\\C题\\附件1.xlsx")
df1=pd.read_excel(xls1)
xls2 = pd.ExcelFile("E:\\数学建模国赛2023选题\\C题\\附件2.xlsx")
df2=pd.read_excel(xls2)
```

```python
# Merge the sales data with the product information based on the common column "单品编码"
merged_data = pd.merge(df2, df1, how='left', on='单品编码')

# Display the first few rows to understand its structure and content after merging
merged_data.head()
```

```python
columns_to_drop = ['单品编码', '分类编码', '销售类型', '是否打折销售']
cleaned_data = merged_data.drop(columns=columns_to_drop)
# 显示前几行以确认删除操作
cleaned_data
```

```python
cleaned_data.to_excel("E:\\数学建模国赛2023选题\\C题\\第一题数据预处理结果(针对品类销售性).xlsx", index=True)
```
