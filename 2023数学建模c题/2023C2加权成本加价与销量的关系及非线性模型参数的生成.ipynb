```python
# Import necessary libraries
import pandas as pd

# Load the wholesale price data from 附件3.xlsx
wholesale_prices_df = pd.read_excel("E:\\数学建模国赛2023选题\\C题\\附件3.xlsx")

# Load the retail price and sales data from df2s.xlsx
retail_sales_df = pd.read_excel("E:\\数学建模国赛2023选题\\C题\\附件2.xlsx")

product_categories_df = pd.read_excel("E:\\数学建模国赛2023选题\\C题\\附件1.xlsx")
```

```python
retail_sales_df = retail_sales_df[(retail_sales_df['销售类型'] == '销售') & (retail_sales_df['是否打折销售'] == '否')]
```

```python
retail_sales_df = retail_sales_df.drop(['销售类型', '是否打折销售'], axis=1)

# Displaying the first few rows of the filtered DataFrame
retail_sales_df.head()
```

```python
# Step 4: Merging the dataframes
# Merging the wholesale prices with retail sales data
merged_wholesale_retail = pd.merge(wholesale_prices_df, retail_sales_df, left_on=['日期', '单品编码'], right_on=['销售日期', '单品编码'], how='inner')

# Merging the above data with product categories
merged_data = pd.merge(merged_wholesale_retail, product_categories_df, on='单品编码', how='left')

# Checking the first few rows of the merged dataframe to make sure it's merged correctly
merged_data.head()
```

```python
# Calculating the weighted average for '日成本加成率(%)', weighted by '销量(千克)'
merged_data['加权日成本加成率(%)'] = merged_data['日成本加成率(%)'] * merged_data['销量(千克)']

# Grouping the data by date and category
grouped_data = merged_data.groupby(['日期', '分类名称']).agg({
    '加权日成本加成率(%)': 'sum',
    '销量(千克)': 'sum'
}).reset_index()

# Calculating the weighted average markup rate for each group
grouped_data['加权日成本加成率(%)'] = grouped_data['加权日成本加成率(%)'] / grouped_data['销量(千克)']

# Checking the first few rows of the grouped data to make sure it's calculated correctly
grouped_data.head()

```

```python
# Step 6: Filtering the data to keep only the records for the months of June, July, and August
# Extracting the month from the date column
grouped_data['月份'] = grouped_data['日期'].dt.month

# Filtering the data
grouped_data = grouped_data[grouped_data['月份'].isin([6, 7, 8])].drop(columns=['月份'])

# Checking the first few rows of the filtered data
grouped_data
```

```python
import pandas as pd

# Assume that the date in '销售日期' starts with a known year, e.g., 2020
# You might need to adjust this based on your actual data
grouped_data['日期'] = pd.to_datetime(grouped_data['日期'].astype(str) + '-2020', errors='coerce')

# Calculate the day of the week and create a new column for it
grouped_data['星期'] = grouped_data['日期'].dt.day_name(locale='zh_CN')

# Display the first few rows of the DataFrame with the new '星期' column
grouped_data
```

```python
# Initialize an empty list to hold the 42 subsets
subsets = []

# Create 7 subsets based on weekdays
for weekday in grouped_data['星期'].unique():
    weekday_data = grouped_data[grouped_data['星期'] == weekday]
    
    # Create 6 subsets based on categories within each weekday subset
    for category in grouped_data['分类名称'].unique():
        category_data = weekday_data[weekday_data['分类名称'] == category]
        subsets.append(category_data)
```
