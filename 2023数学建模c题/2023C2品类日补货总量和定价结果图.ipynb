```python
import pandas as pd

# Load the Excel file
file_path4 = ("E:\\数学建模国赛2023选题\\C题\\附件4.xlsx")

# Check the sheet names first to understand the structure of the Excel file
sheet_names = pd.ExcelFile(file_path4).sheet_names
sheet_names
```

```python
df4 = pd.read_excel(file_path4, sheet_name=sheet_names[0])
```

```python
df4
```

```python
file_path2 = ("E:\\数学建模国赛2023选题\\C题\\第二题数据结果.xlsx")
```

```python
df2 = pd.read_excel(file_path2)
df2
```

```python
df2.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore')
df2
```

```python
merged_data = pd.merge(df2, df4, how='left', left_on='Category', right_on='小分类名称')
merged_data
```

```python
# Calculate Purchase Quantity
merged_data['Purchase Quantity (kg)'] = merged_data['Optimized Sales (kg)'] / (1 - merged_data['平均损耗率(%)_小分类编码_不同值'] / 100)

# Show the first few rows of the final data with the new 'Purchase Quantity (kg)' column
merged_data.head()
```

```python
# Drop the specified columns
final_data = merged_data.drop(columns=['Optimized Sales (kg)', '小分类名称', '平均损耗率(%)_小分类编码_不同值','小分类编码'])
final_data
```

```python
df=final_data
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_theme(style="whitegrid")

# Create the plots
fig, axes = plt.subplots(1, 2, figsize=(18, 8))
plt.rcParams['font.sans-serif'] = ['SimHei']
# Plot for "Optimized Markup Rate (%)"
sns.heatmap(data=df.pivot(index="Day", columns="Category", values="Optimized Markup Rate (%)"), annot=True, fmt=".2f", cmap="coolwarm", ax=axes[0])
axes[0].set_title('成本加价率 (%)')

# Plot for "Optimized Sales (kg)"
sns.heatmap(data=df.pivot(index="Day", columns="Category", values="Purchase Quantity (kg)"), annot=True, fmt=".2f", cmap="coolwarm", ax=axes[1])
axes[1].set_title('日补货总量 (kg))')

plt.tight_layout()
plt.show()
```
