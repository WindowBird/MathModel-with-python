```python
# Import necessary libraries
import pandas as pd

# Load the wholesale price data from 附件3.xlsx
wholesale_prices_df = pd.read_excel("E:\\数学建模国赛2023选题\\C题\\附件3.xlsx")

# Load the retail price and sales data from df2s.xlsx
retail_sales_df = pd.read_excel("E:\\数学建模国赛2023选题\\C题\\附件2.xlsx")

product_categories_df = pd.read_excel("E:\\数学建模国赛2023选题\\C题\\附件1.xlsx")
```

```python
# Step 4: Merging the dataframes
# Merging the wholesale prices with retail sales data
merged_wholesale_retail = pd.merge(wholesale_prices_df, retail_sales_df, left_on=['日期', '单品编码'], right_on=['销售日期', '单品编码'], how='inner')

# Merging the above data with product categories
merged_data = pd.merge(merged_wholesale_retail, product_categories_df, on='单品编码', how='left')

# Checking the first few rows of the merged dataframe to make sure it's merged correctly
merged_data.head()
```

```python
# Step 5: Calculating the markup rate
# Calculating the daily markup rate for each product
merged_data['日成本加成率(%)'] = ((merged_data['销售单价(元/千克)'] - merged_data['批发价格(元/千克)']) / merged_data['批发价格(元/千克)']) * 100

# Grouping the data by date and category to calculate the average daily markup rate for each category
grouped_data = merged_data.groupby(['日期', '分类名称']).agg({
    '日成本加成率(%)': 'mean',
    '销量(千克)': 'sum'
}).reset_index()

# Checking the first few rows of the grouped data to make sure it's calculated correctly
grouped_data

```

```python
# Step 6: Filtering the data to keep only the records for the months of June, July, and August
# Extracting the month from the date column
grouped_data['月份'] = grouped_data['日期'].dt.month

# Filtering the data
grouped_data = grouped_data[grouped_data['月份'].isin([6, 7, 8])].drop(columns=['月份'])

# Checking the first few rows of the filtered data
grouped_data
```

```python
# Importing necessary libraries for date manipulation
from pandas.tseries.offsets import MonthBegin

# Extracting the day of the month from the '日期' column
grouped_data['日期_日'] = grouped_data['日期'].dt.day

# Calculating the mean values for each day of the month and category
grouped_data = grouped_data.groupby(['日期_日', '分类名称']).agg({
    '日成本加成率(%)': 'mean',
    '销量(千克)': 'mean'
}).reset_index()

# Checking the first few rows of the new grouped data
grouped_data
```

```python
import numpy as np

# Define a function to convert Cartesian coordinates to polar coordinates
def cartesian_to_polar(x, y):
    r = np.sqrt(x**2 + y**2)
    theta = np.arctan2(y, x)
    return r, theta

# Selecting data for conversion
x_data = grouped_data['日成本加成率(%)']
y_data = grouped_data['销量(千克)']

# Converting from Cartesian to Polar coordinates
r_data, theta_data = cartesian_to_polar(x_data, y_data)

# Adding the polar coordinates to the DataFrame
grouped_data['r'] = r_data
grouped_data['theta'] = theta_data

# Plotting the data in polar coordinates
plt.figure(figsize=(10, 10))
plt.scatter(theta_data, r_data, c='blue')
plt.title('数据在极坐标下的分布')
plt.xlabel('Theta (角度)')
plt.ylabel('r (距离)')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Taking the natural logarithm of the 'r' data to linearize it
log_r_data = np.log1p(grouped_data['r'])  # Using log1p to handle cases where r=0

# Preparing data for linear regression
X = grouped_data[['日成本加成率(%)', '销量(千克)']]
y = log_r_data

# Fitting a linear model
linear_model = LinearRegression()
linear_model.fit(X, y)

# Getting model parameters
intercept = linear_model.intercept_
coefficients = linear_model.coef_

# Printing the model parameters
intercept, coefficients
```

```python
# Importing 3D plotting module
from mpl_toolkits.mplot3d import Axes3D

# Calculating the predicted log_r values based on the linear model
predicted_log_r = linear_model.predict(X)

# Creating a 3D scatter plot for the actual and predicted log_r values
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for actual log_r values
ax.scatter(X['日成本加成率(%)'], X['销量(千克)'], y, c='b', marker='o', alpha=0.5, label='Actual')

# Scatter plot for predicted log_r values
ax.scatter(X['日成本加成率(%)'], X['销量(千克)'], predicted_log_r, c='r', marker='x', alpha=0.5, label='Predicted')

# Adding labels and title
ax.set_xlabel('日成本加成率(%)')
ax.set_ylabel('销量(千克)')
ax.set_zlabel('ln(r + 1)')
ax.set_title('Actual vs Predicted ln(r + 1)')
ax.legend()
plt.tight_layout()
# Show the plot
plt.show()
```
