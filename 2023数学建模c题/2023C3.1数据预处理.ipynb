```python
import pandas as pd

# Load the Excel file
file_path4 = ("E:\\数学建模国赛2023选题\\C题\\附件4.xlsx")

# Check the sheet names first to understand the structure of the Excel file
sheet_names = pd.ExcelFile(file_path4).sheet_names
sheet_names
```

```python
df4 = pd.read_excel(file_path4, sheet_name=sheet_names[1])
```

```python
df4
```

```python
merged_df = pd.merge(df2, df1, on='单品编码', how='left')
merged_df = pd.merge(merged_df, df3, left_on=['单品编码', '销售日期'], right_on=['单品编码', '日期'], how='left')
merged_df = pd.merge(merged_df, df4, on='单品编码', how='left')
merged_df
```

```python
columns_to_drop = ['扫码销售时间', '分类编码', '分类名称', '日期']
merged_df = merged_df.drop(columns=columns_to_drop, errors='ignore')
```

```python
merged_df
```

```python
# Merge the '单品名称_x' and '单品名称_y' columns into one '单品名称' column
# If '单品名称_x' is missing, use '单品名称_y', otherwise use '单品名称_x'
merged_df['单品名称'] = merged_df.apply(lambda row: row['单品名称_x'] if pd.notna(row['单品名称_x']) else row['单品名称_y'], axis=1)

# Drop the '单品名称_x' and '单品名称_y' columns as they are now redundant
merged_df.drop(columns=['单品名称_x', '单品名称_y'], inplace=True)

# Show the first few rows of the DataFrame with the integrated '单品名称' column
merged_df.head()
```

```python
# Drop the '销售类型' column from the cleaned DataFrame
merged_df.drop(columns=['销售类型'], inplace=True, errors='ignore')

# Show the first few rows of the DataFrame after dropping the '销售类型' column
merged_df.head()
```

```python
# Calculate the proportion of sales that are discounted ('是')
discounted_sales_count = merged_df[merged_df['是否打折销售'] == '是'].shape[0]
total_sales_count = merged_df.shape[0]

# Calculate the proportion
discounted_sales_proportion = discounted_sales_count / total_sales_count * 100

discounted_sales_proportion
```

```python
# Drop the '单品编码' column from the cleaned DataFrame
merged_df.drop(columns=['单品编码'], inplace=True, errors='ignore')

# Show the first few rows of the DataFrame after dropping the '单品编码' column
merged_df
```

```python
#merged_df.to_excel("E:\\数学建模国赛2023选题\\C题\\第三题数据表格.xlsx", index=True)
```

```python
# Calculate the total sales weight and total sales amount for each product
demand_analysis = merged_df.groupby('单品编码').agg({
    '销量(千克)': 'sum',
    '销售单价(元/千克)': 'mean'  # Taking mean as the price might vary
}).reset_index()

# Calculate total sales amount
demand_analysis['总销售额(元)'] = demand_analysis['销量(千克)'] * demand_analysis['销售单价(元/千克)']

# Sort the products by total sales weight and total sales amount
demand_analysis_sorted_by_weight = demand_analysis.sort_values(by='销量(千克)', ascending=False)
demand_analysis_sorted_by_amount = demand_analysis.sort_values(by='总销售额(元)', ascending=False)

# Display the top 10 products by total sales weight and total sales amount
top_10_by_weight = demand_analysis_sorted_by_weight.head(10)
top_10_by_amount = demand_analysis_sorted_by_amount.head(10)

top_10_by_weight
```

```python
top_10_by_amount
```

```python
# Calculate the average wholesale price for each product
supply_analysis = merged_df.groupby('单品编码').agg({
    '批发价格(元/千克)': 'mean'
}).reset_index()

# Merge with the demand analysis data to get a complete view
complete_analysis = pd.merge(demand_analysis, supply_analysis, on='单品编码', how='inner')

# Add a profit margin column (Sales Price - Wholesale Price)
complete_analysis['毛利润(元/千克)'] = complete_analysis['销售单价(元/千克)'] - complete_analysis['批发价格(元/千克)']

# Sort the products by profit margin
complete_analysis_sorted_by_profit = complete_analysis.sort_values(by='毛利润(元/千克)', ascending=False)

# Display the top 10 products by profit margin
top_10_by_profit = complete_analysis_sorted_by_profit.head(10)

top_10_by_profit
```
