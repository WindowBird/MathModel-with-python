```python
import pandas as pd

# Load data
xls = pd.ExcelFile("E:\\数学建模国赛2023选题\\C题\\第三题数据表格.xlsx")
df=pd.read_excel(xls)
df
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore')
df
```

```python
# Calculate the display quantity for each product based on loss rate and sales quantity
# Formula: Display Quantity = Sales Quantity / (1 - Loss Rate)
df['陈列量(千克)'] = df['销量(千克)'] / (1 - df['损耗率(%)'] / 100)


# Show the first few rows of the DataFrame with the calculated display quantity
df.head()
```

```python
# Group the data by '单品名称' and '销售日期', then sum the sales quantity to get daily sales for each product
daily_sales_per_product = df.groupby(['单品名称', '销售日期'])['陈列量(千克)'].sum().reset_index()

# Sort the DataFrame by '单品名称' and '销售日期'
daily_sales_per_product.sort_values(by=['单品名称', '销售日期'], inplace=True)

# Show the first few rows of the daily sales per product DataFrame
daily_sales_per_product
```

```python
# Filter out rows where daily sales per product are less than 2.5 kg
# First, group by '单品名称' and '销售日期' to calculate daily sales
daily_sales = df.groupby(['单品名称', '销售日期'])['陈列量(千克)'].sum().reset_index()

# Identify the products and dates where daily sales are less than 2.5 kg
to_remove = daily_sales[daily_sales['陈列量(千克)'] < 2]

# Remove these rows from the original DataFrame 'df'
df_filtered = df[~((df['单品名称'].isin(to_remove['单品名称'])) & (df['销售日期'].isin(to_remove['销售日期'])))]

# Show the first few rows of the filtered DataFrame 'df_filtered'
df_filtered
```

```python
# Calculate the number of unique products in the filtered DataFrame 'df_filtered'
num_unique_products_filtered = df_filtered['单品名称'].nunique()

num_unique_products_filtered
```
