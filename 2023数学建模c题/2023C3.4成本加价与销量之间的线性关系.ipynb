```python
import pandas as pd

# Load data
xls = pd.ExcelFile("E:\\数学建模国赛2023选题\\C题\\第三题_销量_加权期望销售单价_批发价格_损耗率.xlsx")
df=pd.read_excel(xls)
df
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore')
df
```

```python
# Calculate the markup column using the correct column name: (Expected Selling Price - Wholesale Price) / Wholesale Price * 100
df['成本加价(%)'] = ((df['加权期望销售单价(元/千克)'] - df['批发价格(元/千克)']) / df['批发价格(元/千克)']) * 100

# Show the first few rows to verify
df.head()
```

```python
# Drop the specified columns
df.drop(['加权期望销售单价(元/千克)', '批发价格(元/千克)'], axis=1, inplace=True)

# Show the first few rows to verify
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['axes.unicode_minus'] = False

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Select a few unique products for demonstration
unique_products = df['单品名称'].unique()  # You can adjust this to show more or fewer products

# Generate scatter plots for each selected product
for product in unique_products:
    product_data = df[df['单品名称'] == product]
    
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='销量(千克)', y='成本加价(%)', data=product_data)
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.title(f"Scatter Plot of Sales vs Markup for {product}")
    plt.xlabel("Sales (kg)")
    plt.ylabel("Markup (%)")
    plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Initialize a dictionary to store the regression coefficients for each product
regression_results = {}

# Loop through each unique product to fit a linear regression model
for product in unique_products:
    product_data = df[df['单品名称'] == product]
    
    # Prepare the data
    X = product_data['销量(千克)'].values.reshape(-1, 1)
    y = product_data['成本加价(%)'].values
    
    # Initialize and fit the model
    reg_model = LinearRegression()
    reg_model.fit(X, y)
    
    # Store the regression coefficients
    regression_results[product] = {
        'Intercept': reg_model.intercept_,
        'Slope': reg_model.coef_[0]
    }
    
    # Generate the scatter plot and the regression line
    plt.figure(figsize=(10, 6))
    plt.scatter(X, y, label='Actual Data')
    plt.plot(X, reg_model.predict(X), color='red', label='Regression Line')
    plt.title(f"Linear Regression of Sales vs Markup for {product}")
    plt.xlabel("Sales (kg)")
    plt.ylabel("Markup (%)")
    plt.legend()
    plt.show()

# Display the regression coefficients for each product
regression_results
```

```python
# Convert the regression_results dictionary to a DataFrame
regression_df = pd.DataFrame.from_dict(regression_results, orient='index')

# Show the DataFrame
regression_df
```

```python
"""
\text{成本加价（%）} = \text{Intercept} + (\text{Slope} \times \text{销量（千克）})
Intercept（截距）: 当销量为零时，预测的成本加价值。本质上，它是y轴上的起始点。
Slope（斜率）: 它表示销量每增加一个单位时，成本加价预计会变化多少。斜率是方程中的
"""
```

```python
with pd.ExcelWriter("E:\\数学建模国赛2023选题\\C题\\第三题建模线性函数约束参数.xlsx") as writer:
    regression_df.to_excel(writer, sheet_name='线性函数参数', index=False)
```
