```python
"""粒子群优化（PSO）：

优点：

简单易实现，参数调整较为容易。
由于其并行性，能够同时搜索解空间的多个区域，因此具有较好的全局优化能力。
对于复杂的、非线性的、多模态的、高维度的问题有较好的适应性。
缺点：

对于一些复杂的、需要精确解的问题，粒子群优化可能无法找到最精确的解。
可能会陷入局部最优，尤其是在处理高复杂度、高维度的问题时。
拟牛顿法：

优点：

通常能够快速地收敛到最优解，尤其是在问题是凸的或近似凸的情况下。
不需要提供目标函数的二阶导数，只需要一阶导数，计算成本相对较低。
缺点：

需要计算目标函数的梯度，对于一些不可微或者梯度难以求解的问题，可能无法使用。
通常只能找到局部最优解，不一定能找到全局最优解。
对于非凸、复杂的问题可能表现不佳。"""
```

```python
from scipy.optimize import minimize
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    # x输入粒子位置
    # y 粒子适应度值
    if (x[0]==0)&(x[1]==0):
        y = np.exp((np.cos(2*np.pi*x[0])+np.cos(2*np.pi*x[1]))/2)-2.71289
    else:
        y = np.sin(np.sqrt(x[0]**2+x[1]**2))/np.sqrt(x[0]**2+x[1]**2)+np.exp((np.cos(2*np.pi*x[0])+np.cos(2*np.pi*x[1]))/2)-2.71289
    return -y 

def store_values(x):
    store_values.f_values.append(-func(x))

store_values.f_values = []

x0 = np.array([0, 0])

res = minimize(func, x0, method='BFGS', callback=store_values)

print('Optimized parameters: ', res.x)
print('Minimum function value: ', -res.fun)

plt.plot(store_values.f_values)
plt.xlabel('Iteration')
plt.ylabel('Function value')
plt.show()
```
